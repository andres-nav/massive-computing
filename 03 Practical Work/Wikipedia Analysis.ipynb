```python
def parse_links(document_body):
  data=re.findall(r'\[\[(.+?)\]\]',document_body)
  if (len(data)>0):
    links=[s.lower() for s in data]
  else:
    links=None
  return links
```

```python
parse_links_udf = udf(parse_links,ArrayType(StringType()))
```
